# Data Exploration

```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
```

```python
data = np.load('data.npy')
```

```python
data.shape
```

```python
plt.figure(figsize=(25,2))
plt.plot(data[:250,0,0])
plt.xlim([0,250])
```

```python
fig = plt.figure()
#ax = plt.axes(xlim=(0, 3), ylim=(0, 3))

a = np.random.random((3,3))
im = plt.imshow(a,interpolation='none')
plt.tight_layout()
# initialization function: plot the background of each frame
def init():
    im.set_data(data[0])
    return [im]

# animation function.  This is called sequentially
def animate(i):
    im.set_array(data[i])
    return [im]
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=250, blit=True)
HTML(anim.to_html5_video())
```

```python
flat_data = data.reshape(4500, 1,-1)
print(flat_data.shape)

fig = plt.figure()
#ax = plt.axes(xlim=(0, 3), ylim=(0, 3))

a = np.random.random((1,9))
im = plt.imshow(a,interpolation='none')
plt.tight_layout()
# initialization function: plot the background of each frame
def init():
    im.set_data(flat_data[0])
    return [im]

# animation function.  This is called sequentially
def animate(i):
    im.set_array(flat_data[i])
    return [im]
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=250, blit=True)
HTML(anim.to_html5_video())
```

```python
plt.figure(figsize=(25,2))
plt.plot(np.sum(data[:250].reshape(250, -1),axis=1))
plt.xlim([0,250])
```

```python
plt.figure(figsize=(5,2))
plt.plot(np.sum(data[:18].reshape(18, -1),axis=1))
plt.xlim([0,18])
```

In this we can see there is a periodical change in the data. The changes repeat in 18 data points. This can be used when creating the data splits.

```python
plt.figure(figsize=(25,2))
plt.plot(np.gradient(data[:250].reshape(250, -1),axis=1))
plt.xlim([0,250])
```
